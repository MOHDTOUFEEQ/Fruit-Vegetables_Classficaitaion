```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load the YOLO model with custom weights
model = YOLO('./best_model.pt')  # Ensure path is correct

# Load the image
image_path = './testing6.png'  # Update with correct image path

# Run inference
results = model(image_path)

# Iterate over results and display bounding boxes
for result in results:
    # Convert the result to an image with bounding boxes
    annotated_img = result.plot()

    # Convert from BGR (OpenCV) to RGB (Matplotlib format)
    annotated_img = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)

    # Display the image using Matplotlib
    plt.figure(figsize=(10, 6))
    plt.imshow(annotated_img)
    plt.axis("off")  # Hide axes
    plt.show()

# Optionally, save the output image with detections
cv2.imwrite("./testing_with_detections.jpg", annotated_img)
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load the model
model = YOLO('./best_model.pt')

# Load and run inference on the image
image_path = './testing6.png'
if os.path.exists(image_path):  # Ensure image exists
    results = model(image_path)

    # Convert results to an annotated image
    for result in results:
        annotated_img = result.plot()

        if annotated_img is None:
            print("Error: No output image generated.")
        else:
            # Convert BGR to RGB
            annotated_img = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)

            # Display using Matplotlib
            plt.figure(figsize=(10, 6))
            plt.imshow(annotated_img)
            plt.axis("off")  # Hide axes
            plt.show()
else:
    print("Error: Image file not found!")
```
